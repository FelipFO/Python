```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_wave_propagation(n, iterations, step_size):
    """Simulates wave propagation using Monte Carlo method.

    Args:
        n: Number of random points.
        iterations: Number of iterations.
        step_size: Size of each random step.
    """

    # Initialize points on the initial wavefront (e.g., a line)
    x = np.zeros(n)
    y = np.linspace(0, 1, n)

    # Track the positions of the points over time
    positions = np.zeros((iterations + 1, n, 2))
    positions[0] = np.column_stack((x, y))

    for i in range(iterations):
        # Random walk for each point
        theta = np.random.uniform(-np.pi / 2, np.pi / 2, n)  # Random angles
        dx = step_size * np.cos(theta)
        dy = step_size * np.sin(theta)

        x += dx
        y += dy

        positions[i + 1] = np.column_stack((x, y))

    return positions

# Get parameters from the user
n = int(input("Enter the number of random points: "))
iterations = int(input("Enter the number of iterations: "))
step_size = float(input("Enter the step size: "))

# Simulate wave propagation
positions = simulate_wave_propagation(n, iterations, step_size)

# Visualize the results
plt.figure(figsize=(10, 6))
for i in range(iterations + 1):
    plt.scatter(positions[i, :, 0], positions[i, :, 1], s=5, label=f"Iteration {i}")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Monte Carlo Wave Propagation Simulation")
plt.legend()
plt.grid(True)
plt.show()}
```
